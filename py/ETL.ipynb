```python
# Importing Dependencies
import pandas as pd
import json
from pprint import pprint
from sqlalchemy import create_engine
from config import user, password 

#Making paths to all my JSONs
dogPath = '../Resources/dogs.json'
catPath = '../Resources/cats.json'
rabbitPath = '../Resources/rabbits.json'
birdPath = '../Resources/birds.json'
horsesPath = '../Resources/horses.json'
barnyardPath = '../Resources/barnyard.json'
smallFurryPath = '../Resources/small_furry.json'
scalesPath = '../Resources/scales_fins_others.json'
zipPath = '../Resources/us-zip-code-latitude-and-longitude.json'

paths = [dogPath, catPath, rabbitPath, birdPath, horsesPath, barnyardPath, smallFurryPath, scalesPath]
```

```python
#Establishing a connecting to my SQL database
rds_connection_string = f"{user}:{password}@localhost:5432/pets_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Checking to make sure all tables got read
engine.table_names()
```

```python
#Reading in my JSON to check the data inside
with open(dogPath) as d:
    dogsJson = json.load(d)
    #Setting up a variable to pull all the data from the animals index
    dogs = dogsJson['animals']
    
#Checking it worked
pprint(len(dogs))
```

```python
#Making empty lists to hold the data
age = []
declawed = []
houseTrain = []
shots = []
fixed = []
special = []
mix = []
breedPrim = []
breedSec = []
coat = []
colorPrim = []
colorSec = []
address = []
city = []
state = []
postcode = []
country = []
envCats = []
envKids = []
envDogs = []
gender = []
ids = []
name = []
photo = []
size = []
species = []
status = []
for path in paths:
    
    with open(path) as p:
        animalsJson = json.load(p)
    #Setting up a variable to pull all the data from the animals index
        animals = animalsJson['animals']
        for animal in animals:
    #populating my lists with the JSON data to make my table
            try:
                age.append(animal['age'])
                declawed.append(animal['attributes']['declawed'])
                houseTrain.append(animal['attributes']['house_trained'])
                shots.append(animal['attributes']['shots_current'])
                fixed.append(animal['attributes']['spayed_neutered'])
                special.append(animal['attributes']['special_needs'])
                mix.append(animal['breeds']['mixed'])
                breedPrim.append(animal['breeds']['primary'])
                breedSec.append(animal['breeds']['secondary'])
                coat.append(animal['coat'])
                colorPrim.append(animal['colors']['primary'])
                colorSec.append(animal['colors']['secondary'])
                address.append(animal['contact']['address']['address1'])
                city.append(animal['contact']['address']['city'])
                state.append(animal['contact']['address']['state'])
                postcode.append(animal['contact']['address']['postcode'])
                country.append(animal['contact']['address']['country'])
                envCats.append(animal['environment']['cats'])
                envKids.append(animal['environment']['children'])
                envDogs.append(animal['environment']['dogs'])
                gender.append(animal['gender'])
                ids.append(animal['id'])
                name.append(animal['name'])
                size.append(animal['size'])
                species.append(animal['species'])
                status.append(animal['status'])
                photo.append(animal['photos'][0]['full'])
       
            #There arent photos for every animal so Im putting an except statement here to catch and put none instead     
            except IndexError:
                photo.append("none")
        
        
    
#Printing the first and last items in my loop to make sure they got recorded   
print(len(age))
print(len(address))
print(len(photo))

      

```

```python
#making my dataframe
animalsDf = pd.DataFrame({"id": ids, "name": name, "age": age,"gender": gender,"size": size, "mixed": mix,
                        "primary_breed": breedPrim, "secondary_breed": breedSec, "coat_length": coat, 
                        "primary_color": colorPrim, "secondary_color": colorSec, "declawed": declawed, 
                        "house_trained": houseTrain, "shots_current": shots, "spayed_neutered": fixed,
                        "special_needs": special, "good_with_cats": envCats, "good_with_children": envKids, 
                        "good_with_dogs": envDogs,"species": species,"photo": photo, "address": address,
                        "city": city, "state": state, "postcode": postcode, "country": country,
                        "status": status})

animalsDf
```

```python
#Loading table into SQL
#Taking the Dogs DF and loading it into my table, then checking to see if it worked
animalsDf.to_sql(name='adoptable_pets', con=engine, if_exists='append', index=False)

pd.read_sql_query('select * from adoptable_pets', con=engine).head()
```

```python
#Reading in my JSON tp check the data inside
with open(zipPath) as z:
    zipJson = json.load(z)
    #Setting up a variable to pull all the data from the animals index
    
    
#Checking it worked
print(len(zipJson))
```

```python

zipCodes = []
lats = []
lngs = []

for zips in zipJson:
    zipCodes.append(zips['fields']['zip'])
    lats.append(zips['fields']['latitude'])
    lngs.append(zips['fields']['longitude'])
    
print(len(zipCodes))
```

```python
zipsDf = pd.DataFrame({"postcode": zipCodes, "latitude": lats, "longitude": lngs})

zipsDf.head()
```

```python
#Loading my tables into SQL
#Taking the Zips DF and loading it into my table, then checking to see if it worked
zipsDf.to_sql(name='postcodes', con=engine, if_exists='append', index=False)

pd.read_sql_query('select * from postcodes', con=engine).head()
```
